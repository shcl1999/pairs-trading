```python
import yfinance as yf
from analysis import investigate_cointegration
```

```python
# get data from ING Groep N.V. (ING) and ABN AMRO Bank N.V. (ABN.AS)
abn = yf.download('ABN.AS', start='2022-01-01', end='2023-01-01')
ing = yf.download('ING', start='2022-01-01', end='2023-01-01')
abn = abn[abn.index.isin(ing.index)]
ing = ing[ing.index.isin(abn.index)]
```

```python
nn = yf.download('NN.AS', start='2022-01-01', end='2023-01-01')
asr = yf.download('ASRNL.AS', start='2022-01-01', end='2023-01-01')
nn = nn[nn.index.isin(asr.index)]
asr = asr[asr.index.isin(nn.index)]
```

```python
investigate_cointegration(abn, ing, 0.01, 2, 0.01)
```

```python
investigate_cointegration(nn, asr, 0.1, 8, 0.1)
```
